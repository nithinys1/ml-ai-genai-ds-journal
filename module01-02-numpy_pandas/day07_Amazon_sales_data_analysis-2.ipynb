```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
# Load the dataset
products = pd.read_csv("amazon_products.csv")
# Display the first few rows of the dataset
products.head()
# Display the summary statistics of the dataset
products.describe()
# Display the data types of each column
products.info()
# Check for missing values
products.isnull().sum()
# Display the distribution of product prices
sns.histplot(products['price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
# Display the distribution of product ratings
sns.histplot(products['rating'], bins=20, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
# Display the top 10 most expensive products
top_expensive = products.sort_values(by='price', ascending=False).head(10)
sns.barplot(x='price', y='product_name', data=top_expensive)
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Price')         
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create sample data (no dataset needed)
data = {
    'Product': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Sales': [100, 150, 200, 120, 180, 220],
    'Rating': [4.2, 4.5, 3.8, 4.7, 4.1, 4.6]
}

import pandas as pd
df = pd.DataFrame(data)

# Simple bar plot
sns.barplot(x='Product', y='Sales', data=df, palette='viridis')
plt.title('Product Sales', fontsize=16)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# Simple scatter plot
sns.scatterplot(x='Rating', y='Sales', data=df, s=200, color='red')
plt.title('Rating vs Sales', fontsize=16)
plt.show()

# Simple line plot
sns.lineplot(x='Product', y='Rating', data=df, marker='o')
plt.title('Product Ratings', fontsize=16)
plt.show()
```
