```python
sample_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_name": ["Alice", "Bob", "Charlie", "David"],
    "amount": [100, 150, 200, 250]
})
sample_orders
```

```python
sample_products = pd.DataFrame({
    "order_id": [101, 102, 105, 106],
    "product_name": ["Laptop", "Smartphone", "Tablet", "Headphones"],
    "price": [999.99, 499.99, 299.99, 199.99]
})
sample_products
```

```python
pd.merge(sample_orders, sample_products, on="order_id",  how="inner")
```

```python
pd.merge(sample_orders, sample_products, left_on="order_id", right_on="order_id", how="left")
```

```python
pd.merge(sample_orders, sample_products, left_on="order_id", right_on="order_id", how="right")
```
