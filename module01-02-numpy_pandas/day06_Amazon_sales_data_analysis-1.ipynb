##create new dataframe for testing to check how to remove the duplicate values 


```python
new_test_df = {
    "restaurant_name" : ["Jalsa", "San Churro Cafe", "Jalsa", "San Churro Cafe"],
    "flag": [0, 0, 0, 0],
    "rating": [3.5, 4.0, 3.5, 4.0]
}
new_test_df = pd.DataFrame(new_test_df)
new_test_df
```

```python
new2_df ={
    "people": [2, 4, 2, 4]
}
new2_df = pd.DataFrame(new2_df)
new_test_df = pd.concat([new_test_df, new2_df], axis=1)
new_test_df
```

```python
new_test_df.duplicated()
```

```python
new_test_df.duplicated(keep=False)
```

```python
new_test_df.drop_duplicates(subset=["restaurant_name"], keep='first', inplace=True)
new_test_df
```
